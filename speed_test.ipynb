## Compare the performance of the scikit and the custom decision tree classifiers

Tested on the Wine Quality dataset from UCI Machine Learning Repository. https://archive.ics.uci.edu/ml/datasets/wine+quality

## Import of the dataset

```python
import pandas as pd

data = pd.read_csv('winequality-red.csv')
print("Dataset Classes:")
print(data.iloc[:,-1].unique())
data
```

## Results

```python
x = data.drop(columns="quality")
y = data["quality"]

import time
from my_tree import CustomDecisionTreeClassifier
#Scikit Learn Decision Tree
start = time.time()
for depth in range(1, 20):
    clf = CustomDecisionTreeClassifier(max_depth=depth, min_samples_split=2, min_samples_leaf=1)
    clf.fit(x, y)
end = time.time()
num_of_samples = len(x)
time_taken_full = end-start
print("Time taken by Custom Decision Tree for", num_of_samples, "samples:", time_taken_full)

# Use half of the dataset for fitting
x_half = x.iloc[:len(x)//2]
y_half = y.iloc[:len(y)//2]
start = time.time()
for depth in range(1, 20):
    clf = CustomDecisionTreeClassifier(max_depth=depth, min_samples_split=2, min_samples_leaf=1)
    clf.fit(x_half, y_half)
end = time.time()
num_of_samples = len(x_half)
time_taken_half = end-start
print("Time taken by Custom Decision Tree for", num_of_samples, "samples:", time_taken_half)

# Use 1/4 of the dataset for fitting
x_quarter = x.iloc[:len(x)//4]
y_quarter = y.iloc[:len(y)//4]
start = time.time()
for depth in range(1, 20):
    clf = CustomDecisionTreeClassifier(max_depth=depth, min_samples_split=2, min_samples_leaf=1)
    clf.fit(x_quarter, y_quarter)
end = time.time()
num_of_samples = len(x_quarter)
time_taken_quarter = end-start
print("Time taken by Custom Decision Tree for", num_of_samples, "samples:", time_taken_quarter)

# Calculate ratio of time taken for full dataset and half dataset
ratio = time_taken_full/time_taken_half
print("Ratio of time taken for full dataset and half dataset:", ratio)
# Calculate ratio of time taken for half dataset and quarter dataset
ratio = time_taken_half/time_taken_quarter
print("Ratio of time taken for half dataset and quarter dataset:", ratio)
# Calculate ratio of time taken for full dataset and quarter dataset
ratio = time_taken_full/time_taken_quarter
print("Ratio of time taken for full dataset and quarter dataset:", ratio)
```
